```python
server = "ECOLOGDB2016"
database = "ECOLOGDBVer3"
username = ""
password = ""

con = pyodbc.connect('DRIVER={ODBC Driver 13 for SQL Server}; SERVER='
                     + server +'; DATABASE='+ database + '; Trusted_Connection=yes')
```

```python
query = """
SELECT    ecolog.TRIP_ID,
        MIN(JST) AS start_jst,
        MAX(JST) - MIN(JST) AS time_to_pass ,
        COUNT(ecolog.JST) AS second_to_pass,
        SUM(ecolog.ENERGY_BY_AIR_RESISTANCE) AS air_resistance,
        SUM(ecolog.ENERGY_BY_ROLLING_RESISTANCE) AS rolling_resistance,
        SUM(ABS(ecolog.CONVERT_LOSS)) AS convert_loss,
        SUM(ABS(ecolog.REGENE_LOSS)) AS regene_loss,
        SUM(ecolog.ENERGY_BY_AIR_RESISTANCE) 
        + SUM(ecolog.ENERGY_BY_ROLLING_RESISTANCE)
        + SUM(ABS(ecolog.CONVERT_LOSS))
        + SUM(ABS(ecolog.REGENE_LOSS)) AS loss
FROM    ECOLOG_Doppler AS ecolog
    INNER JOIN SEMANTIC_LINKS AS sl ON ecolog.LINK_ID = sl.LINK_ID
WHERE    sl.SEMANTIC_LINK_ID = 188
    AND    ecolog.DRIVER_ID = 1
    AND ecolog.JST BETWEEN '2018-01-01' AND '2018-12-31'
    AND ecolog.TRIP_DIRECTION = 'outward'
GROUP BY ecolog.TRIP_ID
"""
```

```python
cursor = con.cursor()
cursor.execute(query)
rows = cursor.fetchall()
df = pd.DataFrame(np.array(rows))
```

```python
# CSV作成
header_list = ['trip_id','start_time','time_to_pass','second_to_pass',
               'air_resistance','rolling_resistance','convert_loss','regene_loss','loss']
df.to_csv('ecolog_loss_2018.csv', header=header_list, index=False)
```
